```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ks_df = pd.read_csv("results/Shp2_ks_data.csv", index_col=0)
ks_df
```

```python
plt.figure(figsize=(18,5), dpi=600)
hueorder=["PT-MCMC", "SMC", "PMC"]
sns.boxplot(data=ks_df, x="Param", hue="Method", y="KS", showfliers=True, hue_order=hueorder)
plt.xticks(rotation=45);
plt.ylabel("Two-Way Kolmogorov-Smirnov Statistic")
plt.xlabel("Parameter")
#plt.title(shorthand)
#plt.xticks(plt.gca().get_xticks(), ["$\kappa_A$","$k_2$","$k_3$","$k_4$","$k_5$","dummy"])
plt.axhline(y=0.05, linestyle="--", color="r", lw=2);
```
